# Data Preprocessing (Average)

```
# calculate average sentiment score for each day
# The dataset misses the tweets data from March 29 2020 to March 30 2020
twitter = pd.read_csv("/content/drive/MyDrive/AN6201/corona_tweets_04.csv",header=None)
d = datetime(year=2020,month=3,day=17)
twitter['Date']=d+timedelta(days=4)
twitter.columns=['ID','Sentiment','Date']
twitter.drop(columns=['ID'],inplace=True)
sentiment = twitter.groupby(twitter.Date).mean()
sentiment
```

```
# Sentiment Score from March 21 2020 to March 28 2020
for i in range(5,12):
  if i<=9:
    df = pd.read_csv(f"/content/drive/MyDrive/AN6201/corona_tweets_0{i}.csv",header=None)
    df['Date']=d+timedelta(days=i)
    df.columns=['ID','Sentiment','Date']
    df.drop(columns=['ID'],inplace=True)
    df_score = df.groupby(df.Date).mean()
    sentiment = sentiment.append(df_score)
  else:
    df = pd.read_csv(f"/content/drive/MyDrive/AN6201/corona_tweets_{i}.csv",header=None)
    df['Date']=d+timedelta(days=i)
    df.columns=['ID','Sentiment','Date']
    df.drop(columns=['ID'],inplace=True)
    df_score = df.groupby(df.Date).mean()
    sentiment = sentiment.append(df_score)
sentiment4_11 = sentiment.copy()
sentiment4_11
```

# Data Preprocessing (Negtive Percentage)


```
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
```

```
# calculate average sentiment score for each day
# The dataset misses the tweets data from March 29 2020 to March 30 2020
d = datetime(year=2020,month=3,day=17)
twitter = pd.read_csv("/content/drive/MyDrive/AN6201/corona_tweets_04.csv",header=None)
twitter.columns=['ID','Sentiment']
twitter.drop(columns=['ID'],inplace=True)
sad = twitter[twitter['Sentiment']<-0.05]
percentage = len(sad)/len(twitter)
dataframe = {'Date': [d+timedelta(days=4)], 'percentage':[percentage]}
sentiment = pd.DataFrame(dataframe)
sentiment
```

```
# Sentiment Score from March 21 2020 to March 28 2020
percentage_list_4_11 = []
date_list_4_11 = []
for i in range(5,12):
  if i<=9:
    df = pd.read_csv(f"/content/drive/MyDrive/AN6201/corona_tweets_0{i}.csv",header=None)
    df.columns=['ID','Sentiment']
    df.drop(columns=['ID'],inplace=True)
    sad = df[df['Sentiment']<-0.05]
    percentage = len(sad)/len(df)
    percentage_list_4_11.append(percentage)
    date = d+timedelta(days=i)
    date_list_4_11.append(date)
  else:
    df = pd.read_csv(f"/content/drive/MyDrive/AN6201/corona_tweets_{i}.csv",header=None)
    df.columns=['ID','Sentiment']
    df.drop(columns=['ID'],inplace=True)
    sad = df[df['Sentiment']<-0.05]
    percentage = len(sad)/len(df)
    percentage_list_4_11.append(percentage)
    date = d+timedelta(days=i)
    date_list_4_11.append(date)
dataframe_4_11 = {'Date': date_list_4_11, 'percentage': percentage_list_4_11}
sentiment_4_11 = pd.DataFrame(dataframe_4_11)
sentiment_4_11 = sentiment.append(sentiment_4_11)
sentiment_4_11
```

```
# Sentiment Score from March 30 2020 to June 26 2020
percentage_list_12_256 = []
date_list_12_256 = []
d = datetime(year=2020,month=3,day=18)
for i in range(12,257):
  if i == 223: # corona_tweets_223 lacks all the sentiment scores
    pass
  else: 
    df = pd.read_csv(f"/content/drive/MyDrive/AN6201/corona_tweets_{i}.csv",header=None)
    df.columns=['ID','Sentiment']
    df.drop(columns=['ID'],inplace=True)
    sad = df[df['Sentiment']<-0.05]
    percentage = len(sad)/len(df)
    percentage_list_12_256.append(percentage)
    date = d+timedelta(days=i)
    date_list_12_256.append(date)
dataframe_12_256 = {'Date': date_list_12_256, 'percentage': percentage_list_12_256}
sentiment_12_256 = pd.DataFrame(dataframe_12_256)
sentiment_4_256 = sentiment_4_11.append(sentiment_12_256)
sentiment_4_256
```

```
sentiment_4_256 = sentiment_4_11.append(sentiment_12_256)
sentiment_4_256
```

```
missing_days = {"Date": [datetime(year=2020,month=3,day=29),datetime(year=2020,month=10,day=27)],
                "percentage":[np.nan,np.nan]}
df_missing = pd.DataFrame(data = missing_days)
df_missing
```

```
sentiment = sentiment_4_256.copy()
sentiment = sentiment.append(df_missing)
sentiment.index=sentiment['Date']
sentiment.sort_index(inplace=True)
sentiment.index = range(len(sentiment))
sentiment
```

```
print(sentiment[sentiment['Date']=="2020-03-29"].index) #8
print(sentiment[sentiment['Date']=="2020-10-27"].index) #220
```

```
sentiment['percentage'][8] = (sentiment['percentage'][7]+sentiment['percentage'][9])/2
sentiment['percentage'][220] = (sentiment['percentage'][219]+sentiment['percentage'][221])/2
```
